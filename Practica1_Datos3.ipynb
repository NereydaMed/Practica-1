```
df_dropped = df.dropna(subset=['Winner Pts'])
df_dropped.head()
```

```
df_dropped.shape
```

```
df_copy = df.copy()
df_copy.head()
```

```
df_copy.shape
```

```
df_dropped['Winner Pts'].dtype
```

```
df_dropped['Winner Pts'].mean()
```

```
df_copy['Winner Pts'].fillna(value=np.round(df['Winner Pts'].mean(),decimals=0), inplace=True)
df_copy.agg(['count', 'size', 'nunique'])
```
