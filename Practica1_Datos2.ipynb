```
#DATA TRANSFORM
```

```
df_dropped = df.dropna(subset=['lead_time'])
df_dropped.head()
```

```

df_dropped.shape
```

```
df_copy = df.copy()
df_copy.head()
```

```
df_copy.shape
```

```
df_dropped['adr'].dtype
```

```
df_dropped['adr'].mean()
```

```
df_copy['adr'].fillna(value=np.round(df['adr'].mean(),decimals=0), inplace=True)
df_copy.agg(['count', 'size', 'nunique'])
```

```
#Handling Categorical Data
list(df['meal'].unique())
```

```
df['meal'].replace({"BB":"Bed & Breakfast", 
                    "HB":"Half board",
                    "FB":"Full board", }, inplace=True)
```

```
list(df['meal'].unique())
```

```
keys = list(df['meal'].unique())
vals = range(1,8)
act = dict(zip(keys, vals))
act
```

```
df_copy['meal_dp'] = df['meal'].map(act)
df_copy.head()
```

```
list(df_copy['meal_dp'].unique())
```
